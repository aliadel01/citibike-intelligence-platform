## Download the data

```python
from pathlib import Path
import urllib.request
import pandas as pd
import json
```

```python
url = "https://gbfs.citibikenyc.com/gbfs/en/station_information.json"


output_path = Path("../data/raw/stations.json")
output_path.parent.mkdir(parents=True, exist_ok=True)

with urllib.request.urlopen(url) as response:
    output_path.write_bytes(response.read())    
```

```python
with output_path.open("rb") as f:
    data = json.load(f)

stations = data["data"]["stations"]
df_stations = pd.DataFrame(
    stations,
    columns=["station_id", "name", "lat", "lon", "capacity"]
).rename(columns={"station_id": "id", "lon": "lng"})

processed_path = Path("../data/processed/dim_stations.csv")
processed_path.parent.mkdir(parents=True, exist_ok=True)
df_stations.to_csv(processed_path, index=False)
```
